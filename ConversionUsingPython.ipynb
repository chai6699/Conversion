```python
import tkinter as tk

def cm_to_feet(cm):
    return cm / 30.48

def feet_to_inches(feet):
    return feet * 12

def sqft_to_sqm(sqft):
    return sqft * 0.092903

def sqft_to_hectares(sqft):
    return sqft * 0.0000092903

def sqft_to_acres(sqft):
    return sqft * 0.0000229568

def perform_conversion():
    value = float(entry.get())
    conversion_type = conversion_var.get()
    
    if conversion_type == "cm_to_feet":
        result.set(f"{value} cm = {cm_to_feet(value):.2f} feet")
    elif conversion_type == "feet_to_inches":
        result.set(f"{value} feet = {feet_to_inches(value):.2f} inches")
    elif conversion_type == "sqft_to_sqm":
        result.set(f"{value} sqft = {sqft_to_sqm(value):.2f} sqm")
    elif conversion_type == "sqft_to_hectares":
        result.set(f"{value} sqft = {sqft_to_hectares(value):.6f} hectares")
    elif conversion_type == "sqft_to_acres":
        result.set(f"{value} sqft = {sqft_to_acres(value):.6f} acres")

app = tk.Tk()
app.title("Unit Conversion Tool by Chaithanya:")
app.geometry("500x500+500+250")


conversion_var = tk.StringVar()
conversion_var.set("cm_to_feet")

frame = tk.Frame(master=app, width=100,bg="#12334f")
frame.pack(fill=tk.BOTH, side=tk.LEFT,expand=True)


label = tk.Label(frame, text="Enter a value:",bg="#3597e8",fg="#12334f")
label.grid(row=0, column=0)
label.place(x=150,y=150)

entry = tk.Entry(frame)
entry.grid(row=0, column=1)
entry.place(x=251,y=151)

result = tk.StringVar()
result.set("")
result_label = tk.Label(frame, textvariable=result)
result_label.grid(row=1, column=0, columnspan=2)
result_label.place(x=190,y=190)

convert_button = tk.Button(frame, text="Convert",bg="#3597e8",fg="#12334f", command=perform_conversion)
convert_button.grid(row=2, column=0, columnspan=2)
convert_button.place(x=220,y=250)

conversion_menu = tk.OptionMenu(frame, conversion_var, "cm_to_feet", "feet_to_inches", "sqft_to_sqm", "sqft_to_hectares", "sqft_to_acres")
conversion_menu.grid(row=3, column=0, columnspan=2)
conversion_menu.place(x=190,y=320)

app.mainloop()
```
